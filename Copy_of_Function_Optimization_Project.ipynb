Create python definitions of the following 5 functions and their derivatives:

$f(x) = ln(x^3)$

$f(x) = exp(x^{1/3})$

$f(x) = x^2 sin(x)$

$f(x) = \frac{\sqrt{2}}{sin(x) cos(x)}$

$f(x) = \sigma(x)$ // this is the sigmoid function, important historically for neural nets

````
# LAMBDA SCHOOL
#
# MACHINE LEARNING
#
# MIT LICENSE

# Your code goes here...

import numpy as np
import matplotlib.pyplot as plt


# f(x) = ln(x^3)
def f1(x):
  if (x > 0):
    return np.log(x**3)
  else:
    return None

# f'(x) = 3/x
def df1(x):
  if (x > 0):
    return 3 / x
  else:
    return None
  

# f(x) = exp(x^(1/3))
def f2(x):
  # kludge to handle negative cubed roots without create a complex variable
  power = x**(1/3) if x >= 0 else -((-x)**(1/3))
  return np.exp(power)

# f'(x) = exp(x^(1/3)) / (3 * x^(2/3))
def df2(x):
  # kludge to handle negative cubed roots without create a complex variable
  power = x**(1/3) if x >= 0 else -((-x)**(1/3))
  return np.exp(power) / (3 * power**2)


# f(x) = (x^2)sin(x)
def f3(x):
  return x**2 * np.sin(x)

# f'(x) = (x^2)cos(x) + 2xsin(x) 
def df3(x):
  return x**2 * np.cos(x) + 2 * x * np.sin(x)


# f(x) = (√2)/(sin(x)cos(x))
def f4(x):
  return np.sqrt(2) / (np.sin(x) * np.cos(x))

# f'(x) = (√2)/(cos(x))^2 - (√2)/(sin(x))^2
def df4(x):
  return np.sqrt(2) * ((1 / np.cos(x)**2) - (1 / np.sin(x)**2))


# f(x) = σ(x) = 1 / (1 + exp(-x))
def f5(x):
  return 1 / (1 + np.exp(-x))

#f'(x) = 1 / (2 + exp(x) + exp(-x))
def df5(x):
  return 1 / (2 + np.exp(x) + np.exp(-x))

````

Plot the above five functions each on separate plots.

Where is the derivative of each function equal to zero?


$f(x) = ln(x^3)$

$f'(x) = \frac{3}{x}$

$f'(x)$ is never zero but approaches zero as x goes to infinity

<hr>

$f(x) = exp(x^{1/3})$

$f'(x) = \frac{e^{x^{1/3}}}{3x^{2/3}}$

$f'(x)$ is never zero but approaches zero as x goes to negative infinity

<hr>


$f(x) = x^2 sin(x)$

$f'(x) = x^2 cos(x) + 2 x sin(x)$

$f'(x) = 0 \Rightarrow$ either $x = 0$ or $x cos(x) + 2 sin(x) = 0$

We can try to solve the second equation symbolically or numerically (see farther down)

<hr>


$f(x) = \frac{\sqrt{2}}{sin(x) cos(x)}$

$f'(x) = \frac{\sqrt(2)}{cos^2(x)} - \frac{\sqrt(2)}{sin^2(x)}$

$f'(x) = 0 \Rightarrow$ $cos(x) = \pm sin(x)$

This happens at $x = \pi/4 + {k\pi}/2$, where $k$ is any integer

<hr>


$f(x) = \sigma(x) = \frac{1}{1 + e^{-x}}$

$f'(x) = \frac{1}{2 + e^x + e^{-x}}$

$f'(x)$ is never zero but approaches zero as x goes to $\pm$ infinity

So it looks like the extrema besides at x = 0 are near x = 2.5, 5, and 8

````
# let's write a little numerical solver


# find a root of a function using tangent method
# ah, it's been too long since I've had numerical analysis
# I think this will do the job and should converge pretty quickly
def solve(f, x_init, tol = 0.000000000001, delta = 0.000000001, max_iter = 1000):
  x0 = x_init
  y0 = f(x0)
  if np.abs(y0) < tol:
    return x0
  for _ in range(max_iter):
    slope = (f(x0 + delta) - y0) / delta
    if slope == 0:
      return None
    x0 = x0 - (y0 / slope)
    y0 = f(x0)
    if np.abs(y0) < tol:
      return x0
  return None


# derivative with one 'x' factored out
def dfunc(x):
  return x * np.cos(x) + 2 * np.sin(x)

inits = [2.5, 5, 8]

for init in inits:
  print("Root near x =", init, "is", solve(dfunc, init))
````
